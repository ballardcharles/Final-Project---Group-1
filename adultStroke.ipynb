```python
from sklearn import tree
```

```python
data_dummies = pd.get_dummies(filt2, columns=['gender','ever_married', 'Residence_type','work_type','smoking_status'], drop_first=True)
```

```python
print(data_dummies.shape)
data_dummies.head()
```

```python
# Split data into train and test sets
np.random.seed(42)

X = data_dummies.drop("stroke", axis = 1)
y = data_dummies["stroke"]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
clf.predict(X_test)
```

```python
#dataframe specific to stroke with ID removed
data_dummies2 = data_dummies.drop(['id'], axis=1)
data_dummies2
```

```python
corr_matrix = data_dummies2.corr()
fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.heatmap(corr_matrix,
                 annot=True,
                 linewidths=0.5,
                 fmt=".2f",
                 cmap="vlag");
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier
```

```python
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
predictions = classifier.predict(X_test)
print(f"First 10 Predictions:   {predictions[:10]}")
print(f"First 10 Actual labels: {y_test[:10].tolist()}")
```

```python
pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
```
